## Q1. What is Hypothesis Testing in Statistics?

Answer:-  Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data.  It provides a systematic procedure to test whether a claim (hypothesis) about a population is likely to be true.

---

**Steps in Hypothesis Testing:**
1. **Formulate hypotheses**  
   - Null hypothesis (\(H_{0}\)): No effect or no difference (status quo).  
   - Alternative hypothesis (\(H_{1}\) or \(H_{a}\)): Represents the effect, difference, or claim to be tested.
   
2. **Select significance level (\(\alpha\))**  
   - Common choices: \(\alpha = 0.05\) (5%) or \(\alpha = 0.01\) (1%).  
   - Represents the probability of rejecting \(H_{0}\) when it is true (Type I error).

3. **Choose a suitable test statistic**  
   - Depends on data type and sample size (e.g., *z-test, t-test, chi-square*).

4. **Compute test statistic and p-value**  
   - Compare with critical value or use the p-value method.

5. 5. **Decision**

- If  
$$p\text{-value} \leq \alpha \;\;\Rightarrow\;\; \text{Reject } H_{0}, \; \text{evidence supports } H_{1}$$  

- If  
$$p\text{-value} > \alpha \;\;\Rightarrow\;\; \text{Fail to reject } H_{0}$$


   


---

**Example:**  
A company claims that the average lifetime of its batteries is 500 hours. We want to test this claim.  

- **Null Hypothesis:**  

$$
H_{0} : \mu = 500
$$  

- **Alternative Hypothesis (two-tailed):**  

$$
H_{1} : \mu \neq 500
$$


If a random sample of battery lifetimes is collected and the test yields \(p < 0.05\), we **reject \(H_{0}\)** and conclude that the true mean lifetime is statistically different from 500 hours.

---

**Conclusion & Interpretation:**  
Hypothesis testing ensures decisions are made based on statistical evidence rather than guesswork.  
It balances the risk of errors (Type I and II) while providing a structured way to evaluate claims.

## Q2. What is the Null Hypothesis, and how does it differ from the alternative hypothesis?

Answer-
The null hypothesis is a statistical statement that assumes **no effect, no difference, or no change** in the population.  It represents the "status quo" or the claim we initially assume to be true until there is strong evidence against it.

---

**Definition of Alternative Hypothesis (\(H_{1}\) or \(H_{a}\)):**  
The alternative hypothesis is a statement that **contradicts the null hypothesis**.  
It suggests the presence of an effect, difference, or change, and is what we want to prove with statistical evidence.

---

**Key Differences:**

| Aspect | Null Hypothesis (\(H_{0}\)) | Alternative Hypothesis (\(H_{1}\) / \(H_{a}\)) |
|--------|-----------------------------|----------------------------------------------|
| Meaning | Assumes no effect or no difference | Assumes effect or difference exists |
| Basis | Default or starting assumption | Competing claim against \(H_{0}\) |
| Test Outcome | Retain or reject \(H_{0}\) | Accepted if \(H_{0}\) is rejected |
| Symbol | \(H_{0} : \mu = \mu_{0}\) | \(H_{1} : \mu \neq \mu_{0}, \mu > \mu_{0}, \mu < \mu_{0}\) |

---

**Example:**  
Suppose a manufacturer claims that the mean weight of a product is 1 kg.

- **Null Hypothesis:**  

$$
H_{0} : \mu = 1.0
$$  

- **Alternative Hypothesis (two-sided test):**  

$$
H_{1} : \mu \neq 1.0
$$


If the sample data shows a significant difference from **1.0 kg**, then  

$$
p < \alpha \;\;\Rightarrow\;\; \text{Reject } H_{0} \;\; \text{and support } H_{1}
$$


---

**Conclusion:**  
The null hypothesis provides a **baseline assumption**, while the alternative hypothesis represents the **research claim**.  
Statistical testing helps us decide, based on sample evidence, whether to **retain \(H_{0}\)** or **accept \(H_{1}\)**.

## Q3. Explain the Significance Level in Hypothesis Testing and its role in deciding the outcome of a test.

Answers:-
The **significance level**, denoted as \(\alpha\), is the threshold probability used to decide whether to reject the null hypothesis (\(H_{0}\)).  
It represents the maximum risk we are willing to take of committing a **Type I error** (rejecting \(H_{0}\) when it is actually true).

---

**Common Values of Significance Level (\(\alpha\)):**  

-  
$$
\alpha = 0.05 \;\;\Rightarrow\;\; 5\% \text{ risk of rejecting a true null hypothesis}
$$  

-  
$$
\alpha = 0.01 \;\;\Rightarrow\;\; 1\% \text{ risk (stricter test)}
$$  

-  
$$
\alpha = 0.10 \;\;\Rightarrow\;\; 10\% \text{ risk (more lenient test)}
$$


---

**Role in Hypothesis Testing:**  
1. **Compare \(p\)-value with \(\alpha\):**  

-  
$$
p \leq \alpha \;\;\Rightarrow\;\; \text{Reject } H_{0} \; (\text{evidence supports } H_{1})
$$  

-  
$$
p > \alpha \;\;\Rightarrow\;\; \text{Fail to reject } H_{0} \; (\text{insufficient evidence against it})
$$  


2. **Control of Error:**  

-  
$$
\alpha \; \text{ directly controls the probability of a Type I error.}
$$  

-  
$$
\text{Smaller } \alpha \;\;\Rightarrow\;\; \text{reduces false positives, but may increase false negatives (Type II errors).}
$$


---

**Example:**  
A pharmaceutical company claims a new drug lowers blood pressure.  
We test at significance level  

$$
\alpha = 0.05
$$  

- **Null Hypothesis:**  

$$
H_{0} : \mu = 0 \quad (\text{no effect})
$$  

- **Alternative Hypothesis:**  

$$
H_{1} : \mu < 0 \quad (\text{drug lowers blood pressure})
$$  

If the test yields  

$$
p = 0.03 < 0.05 \;\;\Rightarrow\;\; \text{Reject } H_{0}
$$  

✅ There is sufficient evidence to conclude that the drug is effective.


---

**Conclusion:**  
The significance level (\(\alpha\)) acts as a **decision rule**.  
It helps balance the risks of making incorrect conclusions, ensuring that results are statistically reliable.

## Q4. What are Type I and Type II Errors? Give examples of each.

Answer- When we perform hypothesis testing, two types of incorrect decisions may occur:

1. **Type I Error (\(\alpha\))**  
   - Occurs when we **reject the null hypothesis (\(H_{0}\)) when it is actually true**.  
   - Probability of committing Type I error is equal to the significance level \(\alpha\).  
   - Also called **false positive**.

2. **Type II Error (\(\beta\))**  
   - Occurs when we **fail to reject the null hypothesis (\(H_{0}\)) when it is actually false**.  
   - Probability of committing Type II error is denoted by \(\beta\).  
   - Also called **false negative**.

---

**Error Decision Table:**

| Decision / Reality       | \(H_{0}\) is True | \(H_{0}\) is False |
|--------------------------|-------------------|-------------------|
| **Reject \(H_{0}\)**     | Type I Error (\(\alpha\)) | Correct Decision (Power = \(1-\beta\)) |
| **Fail to Reject \(H_{0}\)** | Correct Decision | Type II Error (\(\beta\)) |

---

**Examples:**

1. **Type I Error Example (False Positive):**  

A company claims a new medicine is effective.  

- **Null Hypothesis:**  

$$
H_{0} : \mu = 0 \quad (\text{no effect})
$$  

- **Alternative Hypothesis:**  

$$
H_{1} : \mu \neq 0 \quad (\text{drug works})
$$  

If we **reject** \(H_{0}\) when in reality the medicine does **not** work, we commit a **Type I error**.  

$$
\text{Risk: Approving an ineffective drug.}
$$


---

2. **Type II Error Example (False Negative):**  
   Using the same drug case:  
   - If the drug truly works but our test **fails to reject \(H_{0}\)** (concludes no effect), we commit a **Type II error**.  
   → Risk: Rejecting a useful medicine.

---

**Conclusion:**  
- **Type I error (\(\alpha\))**: False alarm — seeing an effect when there is none.  
- **Type II error (\(\beta\))**: Missed detection — failing to see an effect that truly exists.  
Both errors are critical, and researchers often try to balance them by choosing appropriate sample size and significance level.

## Q5. What is the difference between a Z-test and a T-test? Explain when to use Each.
Answer-

**1. Z-test:**  
- A statistical test used to determine whether there is a significant difference between sample and population means (or between two sample means) when the **population variance is known** and the sample size is large (\(n > 30\)).  
- Test statistic formula:  

$$
Z = \frac{\bar{X} - \mu}{\tfrac{\sigma}{\sqrt{n}}}
$$




**Where:**  
- $\,\bar{X}\,$ = Sample mean  
- $\,\mu\,$ = Population mean  
- $\,\sigma\,$ = Population standard deviation  
- $\,n\,$ = Sample size  



---

**2. T-test:**  
- Used to test hypotheses about means when the **population variance is unknown** and must be estimated from the sample, especially when sample size is small (\(n \leq 30\)).  
- Test statistic formula:  

$$
t = \frac{\bar{X} - \mu}{\tfrac{s}{\sqrt{n}}}
$$


**Where:**  
- $\,\bar{X}\,$ = Sample mean  
- $\,\mu\,$ = Population mean  
- $\,s\,$ = Sample standard deviation  
- $\,n\,$ = Sample size  


---

**Key Differences Between Z-test and T-test:**  

| Aspect | Z-test | T-test |
|--------|--------|--------|
| Population variance | Known ($\sigma$) | Unknown (estimated by $s$) |
| Sample size | Large ($n > 30$) | Small ($n \leq 30$) |
| Distribution used | Standard Normal Distribution ($N(0,1)$) | Student’s *t*-distribution |
| Accuracy | More accurate for large samples | More accurate for small samples |


**Examples:**

1. **Z-test Example:**  
   A factory claims the mean weight of packed sugar bags is 5 kg with a known variance.  
   If we take a large sample (\(n=100\)), we can use a **Z-test** to check the claim.

2. **T-test Example:**  
   Suppose we want to check whether the average marks of 20 students in a class differ from the national average of 70.  
   Since \(n=20\) is small and variance is unknown, we apply a **T-test**.

---

**Conclusion:**  
- Use a **Z-test** when the sample size is large and population variance is known.  
- Use a **T-test** when the sample size is small and population variance is unknown.  
Both tests serve the same purpose of comparing means but are applied under different conditions.


## Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram. (Include your Python code and output in the code box below.) Hint: Generate random number using random function.

##Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9] (Include your Python code and output in the code box below.)

Answer-

We are given a dataset of product weights (in grams).  
We want to test whether the population mean is equal to **50 grams**.

### Step 1: Formulate Hypotheses
- Null Hypothesis ($H_{0}$): $\mu = 50$  
- Alternative Hypothesis ($H_{1}$): $\mu \neq 50$  


This is a **two-tailed Z-test**.

### Step 2: Z-test Statistic Formula
$$
Z = \frac{\bar{X} - \mu}{\tfrac{\sigma}{\sqrt{n}}}
$$

  
**Where:**  
- $\,\bar{X}\,$ = Sample mean  
- $\,\mu\,$ = Population mean under $H_{0}$  
- $\,\sigma\,$ = Population standard deviation *(if unknown, approximate with sample standard deviation)*  
- $\,n\,$ = Sample size  


### Step 3: Decision Rule
- At significance level \(\alpha = 0.05\), the **critical Z-values** are \(\pm 1.96\).  
- If \(Z \in (-1.96, 1.96)\), we **fail to reject \(H_0\)**.  
- Otherwise, we **reject \(H_0\)**.


```python
import numpy as np
from math import sqrt

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known values
mu = 50  # hypothesized population mean
n = len(sample_data)

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # unbiased estimator

# Compute Z-statistic
Z = (sample_mean - mu) / (sample_std / sqrt(n))

print(f"Sample Size (n): {n}")
print(f"Sample Mean: {sample_mean:.3f}")
print(f"Sample Std Dev: {sample_std:.3f}")
print(f"Z-statistic: {Z:.3f}")

# Critical Z at 5% significance (two-tailed)
Z_critical = 1.96
if abs(Z) < Z_critical:
    print("Result: Fail to reject H0 (No significant difference).")
else:
    print("Result: Reject H0 (Significant difference exists).")
```

### Interpretation of Results
- The computed Z-value will typically fall **within -1.96 and 1.96**.  
- This means the sample mean is **not significantly different** from 50 grams.  
- Hence, we **fail to reject the null hypothesis**.  

**Conclusion:**  
The data supports the claim that the average weight of the product is 50 grams.


## Q8.Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)

Answer-

We want to generate random data from a **Normal distribution** and compute the 95% confidence interval (CI) for the population mean.

### Formula for 95% Confidence Interval:
$$
CI = \bar{X} \;\pm\; Z_{\alpha/2} \times \frac{s}{\sqrt{n}}
$$

**Where:**  


**Where:**  

- $\displaystyle \bar{X}$ = Sample mean  
- $\displaystyle s$ = Sample standard deviation  
- $\displaystyle n$ = Sample size  
- $\displaystyle Z_{\alpha/2}$ = Critical value from Z-table (for 95%, $Z = 1.96$)  



### Steps:
1. Generate 1000 random values from \(N(\mu=100, \sigma=15)\).  
2. Compute the sample mean and standard deviation.  
3. Calculate the 95% confidence interval.  
4. Plot histogram of the data.


```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

# Step 1: Simulate data
np.random.seed(42)   # for reproducibility
mu, sigma, n = 100, 15, 1000
data = np.random.normal(mu, sigma, n)

# Step 2: Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Step 3: 95% Confidence Interval
Z_critical = 1.96  # for 95% CI
margin_of_error = Z_critical * (sample_std / sqrt(n))
CI_lower = sample_mean - margin_of_error
CI_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Std Dev: {sample_std:.2f}")
print(f"95% Confidence Interval: ({CI_lower:.2f}, {CI_upper:.2f})")

# Step 4: Plot histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(CI_lower, color='green', linestyle='--', linewidth=2, label=f"95% CI Lower = {CI_lower:.2f}")
plt.axvline(CI_upper, color='blue', linestyle='--', linewidth=2, label=f"95% CI Upper = {CI_upper:.2f}")
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

Answer-
A **Z-score** represents how many standard deviations a data point is away from the mean.  

### Formula:
$$
Z = \frac{X - \bar{X}}{s}
$$




**Where:**  
- $\displaystyle X$ = Data point  
- $\displaystyle \bar{X}$ = Sample mean  
- $\displaystyle s$ = Sample standard deviation  
  

### Interpretation:
- \(Z = 0\): The value is exactly equal to the mean.  
- \(Z = +1\): The value is **1 standard deviation above** the mean.  
- \(Z = -1\): The value is **1 standard deviation below** the mean.  
- In general, Z-scores help identify **outliers** and standardize datasets for comparison.


```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_zscores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample std deviation
    z_scores = (data - mean) / std_dev
    return z_scores, mean, std_dev

# Example dataset (simulated normal distribution)
np.random.seed(10)
data = np.random.normal(100, 15, 500)

# Calculate Z-scores
z_scores, mean, std_dev = calculate_zscores(data)

print(f"Sample Mean: {mean:.2f}")
print(f"Sample Std Dev: {std_dev:.2f}")
print(f"First 10 Z-scores: {z_scores[:10]}")

# Plot histogram of standardized data
plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', linewidth=2, label="Mean (Z=0)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Conclusion:
- The histogram of Z-scores is centered around 0 (since data is standardized).  
- Most Z-scores lie between -2 and +2, which is expected for normal data.  
- This standardization allows comparison between datasets with different scales.
