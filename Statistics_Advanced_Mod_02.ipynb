## Q1. What is a Random Variable in Probability Theory?

 Answer- A *random variable* is a numerical quantity whose value depends on the outcome of a random experiment.  
It assigns a real number to each outcome in the sample space.  

Random variables are of two main types:  
1. **Discrete Random Variable** – Takes countable values (e.g., number of heads in coin tosses).  
2. **Continuous Random Variable** – Takes uncountable values over an interval (e.g., height of students).  

**Example:**  
- Tossing a fair coin twice: Let `X` = number of heads.  
  Possible outcomes: HH, HT, TH, TT  
  Possible values of `X`: {0, 1, 2} → *Discrete*.  
- Measuring rainfall in a day: Let `Y` = rainfall in mm.  
  Values could be 0, 12.4, 23.56, etc. → *Continuous*.

**Conclusion:**  
Random variables allow us to study and model uncertainty using numerical values, which is essential for applying probability and statistical methods.

## Q4.  What is a binomial distribution, and how is it used in probability?

Answer-A *binomial distribution* models the number of successes in a fixed number of independent trials,  where each trial has only two possible outcomes: **success** or **failure**,  and the probability of success `p` is constant for all trials.

**Conditions for Binomial Distribution:**
1. Fixed number of trials `n`
2. Only two outcomes per trial (success/failure)
3. Probability of success `p` remains constant
4. Trials are independent


---
**Probability Formula:**  
For exactly $k$ successes in $n$ trials:  


**Probability Formula:**  

$$
P(X = k) = \binom{n}{k} \; p^k \; (1-p)^{n-k}
$$  

where:  

$$
\binom{n}{k} = \frac{n!}{k! \; (n-k)!}
$$

**Variable meanings:**  
- $X$ = Random variable representing the number of successes  
- $n$ = Total number of trials  
- $k$ = Number of successes (0 ≤ k ≤ n)  
- $p$ = Probability of success in a single trial  
- $(1-p)$ = Probability of failure in a single trial  
- $n!$ = Factorial of $n$ (product of all positive integers up to $n$)  
---
**Applications:**
- Quality control (number of defective products in a sample)
- Survey results (number of people in favor)
- Coin tossing (number of heads)

**Example:**  
A coin is tossed 5 times. Find the probability of getting exactly 3 heads.  
Here: `n = 5`, `p = 0.5`, `k = 3`.

**Conclusion:**  
The binomial distribution is essential in probability for situations involving repeated independent trials with two possible outcomes.

## Q5. What is the standard normal distribution, and why is it important?
Answer-
The **standard normal distribution** is a special case of the normal distribution  
with a **mean** $\mu = 0$ and a **standard deviation** $\sigma = 1$.  
It is symmetric, bell-shaped, and its total area under the curve is 1.

The probability density function (PDF) is:

$$
f(z) = \frac{1}{\sqrt{2\pi}} \; e^{-\frac{z^2}{2}}
$$

---

**Variable meanings:**
- $z$ = Standardized score (z-score), representing the number of standard deviations from the mean  
- $\mu$ = Mean of the distribution (here $\mu = 0$)  
- $\sigma$ = Standard deviation (here $\sigma = 1$)  
- $e$ = Euler’s number, approximately 2.71828  
- $\pi$ = Pi, approximately 3.14159  

---

**Importance:**
1. **Foundation for Statistical Methods** – Many hypothesis tests, confidence intervals, and quality control methods rely on it.  
2. **Z-Scores** – Any normal distribution can be converted to the standard normal distribution using the transformation:  

$$
z = \frac{x - \mu}{\sigma}
$$  

3. **Table Lookup** – Standard normal tables (z-tables) give probabilities for any range of values.  
4. **Real-World Applications** – Used in fields like finance, psychology, and engineering for decision-making under uncertainty.

---

**Example:**  
If $X \sim N(100, 15)$ (mean 100, standard deviation 15) and $x = 130$, then:

$$
z = \frac{130 - 100}{15} = 2
$$  

This means the score is **2 standard deviations above the mean**.

---

**Conclusion:**  
The standard normal distribution simplifies probability calculations and allows comparisons across different datasets by standardizing values.

## Q6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Answer-  The **Central Limit Theorem (CLT)** states that when we take random samples of size $n$ from any population with mean $\mu$ and standard deviation $\sigma$,  the sampling distribution of the sample mean $\bar{X}$ approaches a normal distribution as $n$ becomes large (usually $n \ge 30$),  regardless of the shape of the original population distribution.

---

**Mathematical Statement:**

If $X_1, X_2, \dots, X_n$ are independent and identically distributed (i.i.d.) with mean $\mu$ and standard deviation $\sigma$, then:

$$
\bar{X} \sim N\left( \mu, \frac{\sigma}{\sqrt{n}} \right) \quad \text{as} \quad n \to \infty
$$

---

**Variable meanings:**
- $X_i$ = $i^{th}$ observation in the sample  
- $\bar{X}$ = Sample mean  
- $\mu$ = Population mean  
- $\sigma$ = Population standard deviation  
- $n$ = Sample size  

---

**Why CLT is Critical:**
1. **Foundation of Inferential Statistics** – Allows the use of normal distribution for hypothesis testing and confidence intervals.  
2. **Works for Non-Normal Populations** – Even if the population is skewed, sample means approximate normality as $n$ increases.  
3. **Simplifies Calculations** – Enables use of $z$ and $t$ statistics in practical problems.

---

**Example:**  
If a population has $\mu = 50$ and $\sigma = 10$, and we take samples of size $n = 36$,  
then the sampling distribution of $\bar{X}$ has:

$$
\mu_{\bar{X}} = \mu = 50
$$

$$
\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}} = \frac{10}{\sqrt{36}} = 1.6667
$$

 This means sample means are much less variable than individual observations.

---

**Conclusion:**  
The CLT is fundamental because it justifies the use of normal-based methods in statistics,  even when dealing with unknown or non-normal populations.


## Q7.What is the significance of confidence intervals in statistical analysis?
Answer-A **confidence interval (CI)** is a range of values, derived from sample data,  that is likely to contain the true population parameter (such as the mean) with a certain level of confidence.

A **(1 − α) × 100%** confidence interval for the mean (when population standard deviation $\sigma$ is known) is given by:

$$
\bar{X} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$

---

**Variable meanings:**
- $\bar{X}$ = Sample mean  
- $\sigma$ = Population standard deviation  
- $n$ = Sample size  
- $Z_{\alpha/2}$ = Z-score corresponding to the desired confidence level  
- $\alpha$ = Significance level (e.g., $\alpha = 0.05$ for 95% CI)  

---

**Interpretation:**  
A 95% confidence interval means that if we took many random samples and computed a CI for each,  about 95% of those intervals would contain the true population mean.

---

**Importance:**
1. **Measures Precision** – Narrower intervals indicate more precise estimates.  
2. **Gives Uncertainty Range** – Communicates the reliability of an estimate.  
3. **Decision-Making** – Used in quality control, medical studies, and surveys to make informed conclusions.

---

**Example:**  
A sample of $n = 36$ has a mean weight $\bar{X} = 50$ kg,population standard deviation $\sigma = 3$ kg,  
and we want a 95% confidence interval.

Steps:
1. Confidence level: 95% → $Z_{0.025} = 1.96$  
2. Margin of error: $E = 1.96 \cdot \frac{3}{\sqrt{36}} = 0.98$  
3. CI: $(50 - 0.98, \; 50 + 0.98) = (49.02, \; 50.98)$

---

**Conclusion:**  
Confidence intervals provide both an estimate and its uncertainty, making statistical results more
informative.

## Q8.What is the concept of expected value in a probability distribution?

Answer- The **expected value (EV)** of a random variable is the long-run average value it takes  over many repetitions of the experiment.  
It represents the *center of gravity* of the probability distribution.

---

**Mathematical Definition:**

For a **discrete random variable** $X$ with values $x_1, x_2, \dots, x_n$  
and corresponding probabilities $p_1, p_2, \dots, p_n$:

$$
E(X) = \sum_{i=1}^{n} x_i \cdot p_i
$$

For a **continuous random variable** with probability density function $f(x)$:

$$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \; dx
$$

---

**Variable meanings:**
- $E(X)$ = Expected value of random variable $X$  
- $x_i$ = Possible value of $X$  
- $p_i$ = Probability of $x_i$ (for discrete case)  
- $f(x)$ = Probability density function (for continuous case)  

---

**Interpretation:**  
The expected value is the mean outcome if the experiment is repeated infinitely many times.  
It does not necessarily have to be a value that $X$ can actually take.

---

**Example:**  
A fair die has faces numbered 1 to 6.  
The probability of each face is $p_i = \frac{1}{6}$.

$$
E(X) = \sum_{i=1}^{6} i \cdot \frac{1}{6}
= \frac{1 + 2 + 3 + 4 + 5 + 6}{6}
= 3.5
$$  

This means the average roll over a large number of throws will be about 3.5.

---

**Conclusion:**  
The expected value is a key measure of central tendency in probability and helps in decision-making, risk analysis, and statistical modeling.

## Q9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.
Answer- A normal distribution is defined by two parameters:
- **Mean ($\mu$)**: Determines the center of the distribution.
- **Standard Deviation ($\sigma$)**: Measures the spread of the distribution.

We will:
1. Generate 1000 random numbers from $N(\mu=50, \sigma=5)$.
2. Compute the sample mean and standard deviation.
3. Visualize the data using a histogram.

---

**Expected properties:**
- The generated sample mean should be close to 50.
- The generated sample standard deviation should be close to 5.






```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 50       # mean
sigma = 5     # standard deviation
n_samples = 1000

# Step 1: Generate random numbers
data = np.random.normal(mu, sigma, n_samples)

# Step 2: Compute mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=0)  # population SD

# Step 3: Print results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")

# Step 4: Plot histogram
plt.figure(figsize=(8,4))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title(f"Histogram of N({mu}, {sigma}²) Sample")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

## Q10. Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

- Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
- Write the Python code to compute the mean sales and its confidence interval.

Answer-The **Central Limit Theorem (CLT)** states that if we take sufficiently large random samples from a population,  
the sampling distribution of the sample mean will be approximately normal, regardless of the shape of the original population.

In this problem:
- **Population:** Daily sales data (20 observations)
- **Goal:** Estimate the average sales and construct a 95% confidence interval (CI)
- **Assumption:** Sales data is independent and identically distributed (i.i.d.)

---

**Formula for CI (when $\sigma$ is unknown):**

$$
\bar{X} \pm t_{\alpha/2, \, df} \cdot \frac{s}{\sqrt{n}}
$$

**Where:**
- $\bar{X}$ = Sample mean  
- $s$ = Sample standard deviation  
- $n$ = Sample size  
- $t_{\alpha/2, \, df}$ = Critical t-value with $df = n - 1$ degrees of freedom  
- $\alpha$ = Significance level (0.05 for 95% CI)  

---

**Steps to Apply:**
1. Calculate the sample mean $\bar{X}$ and standard deviation $s$.
2. Determine the t-critical value for 95% confidence and $n - 1$ degrees of freedom.
3. Calculate the **margin of error**.
4. Construct the **confidence interval**.

---


```python
import numpy as np
from scipy import stats
import math

# Sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Step 1: Mean and standard deviation
x_bar = np.mean(daily_sales)
s = np.std(daily_sales, ddof=1)  # sample standard deviation
n = len(daily_sales)

# Step 2: t-critical value for 95% CI
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 3: Margin of error
margin_error = t_critical * (s / math.sqrt(n))

# Step 4: Confidence interval
lower_bound = x_bar - margin_error
upper_bound = x_bar + margin_error

# Output
print(f"Sample Mean Sales: {x_bar:.2f}")
print(f"Sample Standard Deviation: {s:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

- **Interpretation:**
The calculated 95% confidence interval provides a range within which we expect the **true average daily sales**  for the retail company to lie, with 95% certainty.

- **Conclusion:**
By applying the CLT, even with a moderate sample size, we can estimate the population mean and quantify the uncertainty  of our estimate using a confidence interval. This helps management in planning, forecasting, and decision-making.
